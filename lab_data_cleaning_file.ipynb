```python
import data_cleaning
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

insurance_data = pd.read_csv(url)

insurance_data = data_cleaning.cleaning_columns(insurance_data)
insurance_data = data_cleaning.cleaning_invalid_values(insurance_data)
insurance_data = data_cleaning.formatting_dtypes(insurance_data)
insurance_data = data_cleaning.drop_duplicates(insurance_data)
insurance_data = data_cleaning.null_value_cleaning(insurance_data)

print(insurance_data)
```
